```python
from transformers import pipeline

n = pipeline("text-to-speech",
                    model="kakao-enterprise/vits-ljs")
```

```python
text = """
A team of scientists from NASA's Jet Propulsion Laboratory (JPL), in collaboration with researchers from the European Space Agency (ESA), has successfully deployed a new generation of satellites equipped with advanced sensors to monitor and track the movements of glaciers in Antarctica.\
 These satellites provide high-resolution images and data that are crucial for understanding the dynamics of ice flow and predicting future sea-level rise. The project marks a significant advancement in our ability to study the impacts of climate change on polar ice sheets.
"""
```

```python
!apt-get update
!apt-get install espeak-ng
```

```python
n_text = n(text)
```
