```python
# Logging is the process of recording events, messages, or data points that occur during the execution of a program
# logging module to output information, warnings, errors, and other messages from the transformers library

from transformers.utils import logging
logging.set_verbosity_error()
```

```python
from transformers import pipeline
```

```python
# very small model
chatbot = pipeline(task="conversational",
                   model="facebook/blenderbot-400M-distill")
```

```python
user_message="""
what should I play now
"""
```

```python
from transformers import Conversation
```

```python
conversation = Conversation(user_message)
```

```python
print(conversation)
```

```python
conversation = chatbot(conversation)
```

```python
print(conversation)
```

```python
print(chatbot(Conversation("What do you recommend?")))
```

This is how we can continue chatting.
This is a pretty basic model.
The above model doesn't have memory of any prior conversations.
To include prior conversations, we have to add message to include the previous chat history

```python
conversation.add_message(
    {"role": "user",
     "content": """
I want to play some outdoor games like cricket or football. What do you recommend?

"""
    })
```

```python
print(conversation)
```

```python
conversation = chatbot(conversation)

print(conversation)
```

We can continue chatting using the above method
