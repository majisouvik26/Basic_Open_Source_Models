```python
from transformers.utils import logging
logging.set_verbosity_error()
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
s1 = ['Hitesh is my Friend', 'He is good at academics', 'He is a very good football player']
```

```python
#embeddings
em1 = model.encode(s1, convert_to_tensor=True)
print(em1)
```

```python
s2 = ['Harry Kane is from England', 'He plays football for Bayern Munich', 'This year, Bayern Munich could not win Bundesliga']
```

```python
em2= model.encode(s2, convert_to_tensor=True)
```

```python
print(em2)
```

Now let's calculate cosine similarity between two sentences

```python
# Similarity check
from sentence_transformers import util
```

```python
cosine_scores = util.cos_sim(em1,em2)
```

```python
print(cosine_scores)
```

```python
for i in range(len(s1)):
    print("{} \t\t {} \t\t Score: {:.4f}".format(s1[i],
                                                 s2[i],
                                                 cosine_scores[i][i]))
```
