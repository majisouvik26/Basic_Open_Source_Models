```python
from transformers import pipeline
```

```python
get_completion = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
```

```python
def summarize(input):
    output = get_completion(input)
    return output[0]['summary_text']
```

```python
text = ('''The Mumbai Indians are a franchise cricket team that has etched its name in the annals of the Indian Premier League (IPL) with their remarkable success. Owned by the business magnate Mukesh Ambani, this team adorned in iconic blue and gold colors, featuring the iconic Maharashtrian Hindu sculpted male torso crest, has become a symbol of cricketing excellence.
Led by the charismatic Rohit Sharma, the Mumbai Indians boast a star-studded lineup comprising some of the biggest names in Indian and international cricket. The Wankhede Stadium, nestled in the heart of Mumbai, serves as their fortress, where their passionate fan base creates an electrifying atmosphere during matches. The team's strategic player acquisitions, astute team management, and nurturing of young talent through their mentorship program have been the driving forces behind their sustained success.
Not just on the field, but the Mumbai Indians have also been at the forefront of promoting cricket as a platform for social causes, actively engaging in various community initiatives. Their commercial success and global reach have transformed them into one of the most valuable and recognizable sports brands in India. With an unwavering spirit, stellar performances, and a loyal fan base, the Mumbai Indians continue to be one of the most formidable forces in the IPL, leaving an indelible mark on the cricketing world.''')
# generated by chatgpt-3.5
```

```python
get_completion(text)
```
