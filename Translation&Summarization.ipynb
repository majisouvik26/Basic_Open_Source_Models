```python
from transformers import pipeline
import torch
```

```python
translator = pipeline(task="translation", model="facebook/nllb-200-distilled-600M", torch_dtype=torch.bfloat16)
```

```python
text = """\
I have a very good friend. His name is Hitesh.\ He is a very good football player.
"""
```

```python
text_translated = translator(text, src_lang="eng_Latn", tgt_lang="ben_Beng")
```

```python
print(text_translated)
```
